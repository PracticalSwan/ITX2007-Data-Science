# Question 1

```python
foodsdata = pd.read_csv("foods_data.csv", index_col=0)
foodsmg = pd.read_csv('food_mg.csv', index_col=0)
```

```python
foodsdata.isna().sum()['FoodGroup']
```

```python
foodsdata.isna().sum()
```

```python
missing_count = foodsdata.isna().sum()
threshold = 200
columns_to_drop = missing_count[missing_count > threshold].index.tolist()
foodsdata = foodsdata.drop(columns=columns_to_drop)
print(f"Dropped columns: {columns_to_drop}")
foodsdata.isna().sum()
```

```python
foodsdata.isna().sum()
```

```python
foodsdata = foodsdata.dropna()
foodsdata
```

```python
foodsdata.isna().sum()
```

```python
foodsdata.groupby('FoodGroup').mean(numeric_only=True)
```

```python
merged_data = foodsdata.merge(foodsmg, left_index=True, right_index=True, how='outer')
merged_data
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

columns_to_plot = ['Energy_kcal', 'Protein_g', 'Fiber_g', 'Calcium_mg', 'Iron_mg', 'Zinc_mg']
correlation_matrix = merged_data[columns_to_plot].corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```
