## How many clusters do you see?

Looking at the scatter plot, I can see **3 distinct clusters**:

1. **Cluster 1**: Bottom-left region (negative x, negative y values)
2. **Cluster 2**: Top-right region (positive x, positive y values)  
3. **Cluster 3**: Middle-left region (negative x, near zero or slightly positive y values)

```python
# Import KMeans
from sklearn.cluster import KMeans
import pandas as pd
import numpy as np

# Load data from new_points.csv
new_points = pd.read_csv('new_points.csv', index_col=0)

# Convert to NumPy array
new_points_array = new_points.values

# Create a KMeans instance with 3 clusters
model = KMeans(n_clusters=3)

# Fit the model to the array of points
model.fit(new_points_array)

# Predict the cluster labels
labels = model.predict(new_points_array)

# Print the cluster labels
print(labels)
```

```python
import matplotlib.pyplot as plt

# Assign column 0 of new_points to xs, and column 1 to ys
xs = new_points_array[:, 0]
ys = new_points_array[:, 1]

# Make a scatter plot of xs and ys, colored by cluster labels
plt.scatter(xs, ys, c=labels, alpha=0.5)

# Compute the coordinates of the centroids
centroids = model.cluster_centers_

# Assign centroid coordinates
centroids_x = centroids[:, 0]
centroids_y = centroids[:, 1]

# Make a scatter plot of centroids as diamond markers
plt.scatter(centroids_x, centroids_y, marker='D', s=50, c='blue', label='Centroids')

# Add labels and title
plt.xlabel('x coordinate')
plt.ylabel('y coordinate')
plt.title('K-Means Clustering with Centroids')
plt.legend()
plt.show()
```

## Elbow Method for Optimal Clusters (Seeds Dataset)

In this exercise, we'll use the **elbow method** to determine the optimal number of clusters for the grain (seeds) dataset. The seeds dataset contains measurements of wheat grains with 7 features (area, perimeter, compactness, length, width, asymmetry, groove length).

```python
import pandas as pd
import numpy as np

# Load the seeds dataset
seeds_df = pd.read_csv('seeds.csv', header=None)

# Display the first few rows and shape
print("First 5 rows of seeds dataset:")
print(seeds_df.head())
print(f"\nDataset shape: {seeds_df.shape}")
print(f"Columns: {seeds_df.columns.tolist()}")
```

```python
# Convert to NumPy array (exclude the last column which is the label)
samples = seeds_df.iloc[:, :-1].values

print(f"Samples array shape: {samples.shape}")
print(f"Number of features: {samples.shape[1]}")
print(f"Number of samples: {samples.shape[0]}")
print(f"\nFirst 3 samples:\n{samples[:3]}")
```

```python
from sklearn.cluster import KMeans

# Elbow method: Calculate inertia for k from 1 to 10
ks = range(1, 11)
inertias = []

for k in ks:
    # Create a KMeans instance with k clusters
    model = KMeans(n_clusters=k, random_state=42)
    
    # Fit the model to the grain data samples
    model.fit(samples)
    
    # Append the value of the inertia_ attribute to the list
    inertias.append(model.inertia_)
    
    print(f"k={k}: inertia = {model.inertia_:.2f}")

print(f"\nInertias: {inertias}")
```

```python
import matplotlib.pyplot as plt

# Plot the elbow curve
plt.figure(figsize=(10, 6))
plt.plot(ks, inertias, '-o', color='steelblue', markersize=8)
plt.xlabel('Number of clusters (k)')
plt.ylabel('Inertia')
plt.title('Elbow Method for Optimal Number of Clusters\n(Seeds Dataset)')
plt.xticks(ks)
plt.grid(True, alpha=0.3)

# Add annotation for the elbow point
plt.axvline(x=3, color='red', linestyle='--', alpha=0.5, label='Elbow point (k=3)')
plt.legend()

plt.show()

print("\n## Analysis:")
print("Looking at the elbow plot, we can see the curve 'bends' or 'elbows' around k=3.")
print("This suggests that **3 clusters** is a good choice for the seeds dataset.")
print("\nThe seeds dataset actually contains 3 varieties of wheat (Kama, Rosa, Canadian),")
print("which confirms that k=3 is the optimal number of clusters!")
```

```python
import pandas as pd
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans

# Load data from fish.csv WITHOUT index_col to preserve species as a column
fish_df = pd.read_csv('fish.csv')

# Display the original data
print("Original fish data:")
print(fish_df.head())
print(f"\nDataFrame shape: {fish_df.shape}")
print(f"Columns: {fish_df.columns.tolist()}")

# The first column contains species - let's rename it properly
# Looking at the data, column '1→Bream' should be 'Species'
# Let's reload with proper handling

# Reload with proper column names - the CSV has no header, so we need to handle it
fish_df = pd.read_csv('fish.csv', header=None)
fish_df.columns = ['Species'] + [f'feature_{i}' for i in range(1, 7)]

print("\nData with proper column names:")
print(fish_df.head())
print(f"\nDataFrame shape: {fish_df.shape}")
print(f"Columns: {fish_df.columns.tolist()}")

# Drop the species column to get numeric data only
fish_numeric = fish_df.drop('Species', axis=1)

print("\nNumeric data for clustering:")
print(fish_numeric.head())
print(f"\nNumeric DataFrame shape: {fish_numeric.shape}")

# Create an instance of StandardScaler called scaler
scaler = StandardScaler()

# Create an instance of KMeans with 4 clusters called kmeans
kmeans = KMeans(n_clusters=4)

# Create a pipeline called pipeline that chains scaler and kmeans
pipeline = make_pipeline(scaler, kmeans)

# Fit the pipeline to the fish numeric data
pipeline.fit(fish_numeric)

# Get the cluster labels using .predict() method
labels = pipeline.predict(fish_numeric)

print(f"\nCluster labels: {labels}")
print(f"Unique clusters: {set(labels)}")
```

```python
# Extract the species column from fish_df
species = fish_df['Species']

# Create a DataFrame df with two columns named 'labels' and 'species'
df = pd.DataFrame({'labels': labels, 'species': species})

print("DataFrame with labels and species:")
print(df.head(20))
print(f"\nDataFrame shape: {df.shape}")

# Create a cross-tabulation ct of df['labels'] and df['species']
ct = pd.crosstab(df['labels'], df['species'])

print("\n" + "="*50)
print("Cross-Tabulation of Cluster Labels vs Species")
print("="*50)
print(ct)
print("\n" + "="*50)
print("Analysis:")
print("="*50)
print("The cross-tabulation shows how the KMeans clustering grouped the fish samples.")
print("Each row represents a cluster (0-3), and each column represents a fish species.")
print("The values show how many samples of each species were assigned to each cluster.")
print("\nIdeally, each cluster would predominantly contain samples from only one species,")
print("indicating that the clustering algorithm successfully discovered the species groupings.")
```
