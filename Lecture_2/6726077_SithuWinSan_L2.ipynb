```python
# Weekly Assignment
df = pd.read_csv('L2_Superstore_Sales.csv', index_col=0)
df['Order_Date'] = pd.to_datetime(df['Order_Date'], errors='coerce')
df['Ship_Date'] = pd.to_datetime(df['Ship_Date'], errors='coerce')
df['Order_Year'] = df['Order_Date'].dt.year
print(df.info())
```

```python
# Total sales between year 2015-2018
total_sales = df[(df['Order_Year'] >= 2015) & (df['Order_Year'] <= 2018)]['Sales'].sum()
print(total_sales)

# Bar chart - Sales by Year
year_sales = df[df['Order_Year'].between(2015, 2018)].groupby('Order_Year')['Sales'].sum()
year_sales.plot(kind='bar')
plt.title('Sales by Year')
plt.xlabel('Year')
plt.ylabel('Total Sales (USD)')
plt.show()
```

```python
# Category with highest sales amount (no. of orders)
cat_orders = df['Category'].value_counts()
print(cat_orders)
highest_orders = cat_orders.idxmax()
print(highest_orders)

# Bar chart - Orders by Category
cat_orders.plot(kind='bar')
plt.title('Orders by Category')
plt.xlabel('Category')
plt.ylabel('Number of Orders')
plt.show()
```

```python
# Category with highest sales volume (sales in USD)
cat_sales = df.groupby('Category')['Sales'].sum()
print(cat_sales)
highest_sales = cat_sales.idxmax()
print(highest_sales)

# Bar chart - Sales Volume by Category
cat_sales.plot(kind='bar')
plt.title('Sales Volume by Category')
plt.xlabel('Category')
plt.ylabel('Total Sales (USD)')
plt.show()
```

```python
# Sorted Subcategory Sales
subcat_sales = df.groupby('Sub_Category')['Sales'].sum().sort_values(ascending=False)
subcat_sales.plot(kind='bar')
plt.title('Sales by Subcategory')
plt.xlabel('Subcategory')
plt.ylabel('Total Sales (USD)')
plt.xticks(rotation=90)
plt.show()
```

```python
# Sales by Region
region_sales = df.groupby('Region')['Sales'].sum().sort_values(ascending=False)
region_sales.plot(kind='bar')
plt.title('Sales by Region')
plt.xlabel('Region')
plt.ylabel('Total Sales (USD)')
plt.show()
```
