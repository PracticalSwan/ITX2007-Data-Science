```python
homelessness = pd.read_csv('homelessness.csv')
print(homelessness.describe())
print(homelessness.shape)
print(homelessness.values)
print(homelessness.columns)
print(homelessness.index)
```

```python
homelessness_ind = homelessness.sort_values('individuals')
print(homelessness_ind.head())
```

```python
homelessness_fam = homelessness.sort_values('family_members', ascending=False)
print(homelessness_fam.head())
```

```python
homelessness_reg_fam = homelessness.sort_values(['region', 'family_members'], ascending=[True, False])
print(homelessness_reg_fam.head())
```

```python
state_fam = homelessness[['state', 'family_members']]
print(state_fam.head())
```

```python
ind_gt_10k = homelessness[homelessness['individuals'] > 10000]
print(ind_gt_10k)
```

```python
mountain_reg = homelessness[homelessness['region'] == 'Mountain']
print(mountain_reg)
```

```python
fam_lt_1k_pac = homelessness[(homelessness['family_members'] < 1000) & (homelessness['region'] == 'Pacific')]
print(fam_lt_1k_pac)
```

```python
south_mid_atlantic = homelessness[homelessness['region'].isin(['South Atlantic','Mid-Atlantic'])]
print(south_mid_atlantic)
```

```python
canu = ['California', 'Nevada', 'Utah', 'Arizona']
mojave_homelessness = homelessness[homelessness['state'].isin(canu)]
print(mojave_homelessness)
```

```python
# Adding new columns
homelessness['total'] = homelessness['individuals'] + homelessness['family_members']
print(homelessness[['state', 'individuals', 'family_members', 'total']].head())
```

```python
# Aggregation functions
print(homelessness['individuals'].mean())
print(homelessness['individuals'].median())
print(homelessness['family_members'].min())
print(homelessness['family_members'].max())
```

```python
# Custom aggregation function
def pct30(column):
    return column.quantile(0.3)

print(homelessness['individuals'].agg(pct30))
print(homelessness[['individuals', 'family_members']].agg(pct30))
```

```python
# Cumulative sum
homelessness['cum_individuals'] = homelessness['individuals'].cumsum()
print(homelessness[['state', 'individuals', 'cum_individuals']].head())
```

```python
# Drop duplicates
print(homelessness.drop_duplicates(subset='region').head())
print(homelessness['region'].value_counts())
print(homelessness['region'].value_counts(normalize=True))
```

```python
# GroupBy operations
print(homelessness.groupby('region')['individuals'].mean())
print(homelessness.groupby('region')['individuals'].agg(['min', 'max', 'sum']))
print(homelessness.groupby('region')[['individuals', 'family_members']].mean())
```

```python
# GroupBy with numpy functions
import numpy as np
stats_by_region = homelessness.groupby('region')['individuals'].agg([min, max, np.mean, np.median])
print(stats_by_region)
```

```python
# Set index
homelessness_ind = homelessness.set_index('state')
print(homelessness_ind.head())
print(homelessness_ind.reset_index().head())
```

```python
# Multi-level index
homelessness_multi = homelessness.set_index(['region', 'state'])
print(homelessness_multi.head())
print(homelessness_multi.loc['Mountain'])
```

```python
# Load and visualize avocados data
import matplotlib.pyplot as plt
avocados = pd.read_csv('avocados.csv')
print(avocados.head())
```

```python
# Bar plot - nb_sold by type
nb_sold_by_type = avocados.groupby('type')['nb_sold'].sum()
nb_sold_by_type.plot(kind='bar')
plt.ylabel('Total Volume Sold')
plt.title('Avocado Sales by Type')
plt.show()
```

```python
# Line plot - Sales over time
avocados['date'] = pd.to_datetime(avocados['date'])
sales_by_date = avocados.groupby('date')['nb_sold'].sum()
sales_by_date.plot(kind='line', rot=45)
plt.ylabel('Total Volume')
plt.title('Avocado Sales Over Time')
plt.show()
```

```python
# Scatter plot
avocados.plot(x='nb_sold', y='avg_price', kind='scatter', title='Volume vs Average Price')
plt.show()
```

```python
# Histogram - comparing two types
avocados[avocados['type'] == 'conventional']['avg_price'].plot(kind='hist', alpha=0.7)
avocados[avocados['type'] == 'organic']['avg_price'].plot(kind='hist', alpha=0.6)
plt.legend(['conventional', 'organic'])
plt.xlabel('Average Price')
plt.title('Price Distribution by Type')
plt.show()
```
