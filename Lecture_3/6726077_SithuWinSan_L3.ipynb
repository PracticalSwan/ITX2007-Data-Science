```python
import pandas as pd
homelessness = pd.read_csv('homelessness.csv')
```

```python
# 1. Sort by number of homeless individuals (ascending)
homelessness_ind = homelessness.sort_values('individuals')
homelessness_ind.head()
```

```python
# 2. Sort by number of family_members (descending)
homelessness_fam = homelessness.sort_values('family_members', ascending=False)
homelessness_fam.head()
```

```python
# 3. Sort first by region (asc) then by family_members (desc)
homelessness_reg_fam = homelessness.sort_values(['region', 'family_members'], ascending=[True, False])
homelessness_reg_fam.head()
```

```python
# 4. Create state_fam DataFrame with only state and family_members
state_fam = homelessness[['state', 'family_members']]
state_fam.head()
```

```python
# 5. Filter homelessness for individuals > 10000
ind_gt_10k = homelessness[homelessness['individuals'] > 10000]
ind_gt_10k
```

```python
# 6. Filter homelessness for region == 'Mountain'
mountain_reg = homelessness[homelessness['region'] == 'Mountain']
mountain_reg
```

```python
# 7. Filter homelessness for family_members < 1000 and region == 'Pacific'
fam_lt_1k_pac = homelessness[(homelessness['family_members'] < 1000) & (homelessness['region'] == 'Pacific')]
fam_lt_1k_pac
```

```python
# 8. Filter homelessness for region in ['South Atlantic', 'Mid-Atlantic']
south_mid_atlantic = homelessness[homelessness['region'].isin(['South Atlantic', 'Mid-Atlantic'])]
south_mid_atlantic
```

```python
# 9. Filter homelessness for states in Mojave list 'canu'
canu = ['California', 'Nevada', 'Utah', 'Arizona']
mojave_homelessness = homelessness[homelessness['state'].isin(canu)]
mojave_homelessness
```

```python
# 10. Add total column to homelessness
homelessness['total'] = homelessness['individuals'] + homelessness['family_members']
homelessness.shape
homelessness
```

```python
# 11. Add p_individuals column (proportion of homeless who are individuals)
homelessness['p_individuals'] = homelessness['individuals'] / homelessness['total']
homelessness
```

```python
# 12. Add 'indiv_per_10k' (individuals per 10k population)
homelessness['indiv_per_10k'] = 10000 * homelessness['individuals'] / homelessness['state_pop']

# Subset rows where indiv_per_10k > 20
high_homelessness = homelessness[homelessness['indiv_per_10k'] > 20]

# Sort descending by indiv_per_10k
high_homelessness_srt = high_homelessness.sort_values('indiv_per_10k', ascending=False)

# Select only state and indiv_per_10k
result = high_homelessness_srt[['state', 'indiv_per_10k']]
result
```

```python
# Create a DataFrame, sales_1_1, which contains the sales data for department 1 of store 1.
sales_1_1 = sales[(sales['department'] == 1) & (sales['store'] == 1)]
sales_1_1
```

```python
# Sort sales_1_1 by date
sales_1_1 = sales_1_1.sort_values('date', ascending = True)

# Get the cumulative sum of weekly_sales, add as cum_weekly_sales col
sales_1_1['cum_weekly_sales'] = sales_1_1['weekly_sales'].cumsum()

# Get the cumulative max of weekly_sales, add as cum_max_sales col
sales_1_1['cum_max_sales'] = sales_1_1['weekly_sales'].cummax()

# See the columns you calculated
print(sales_1_1[["date", "weekly_sales", "cum_weekly_sales", "cum_max_sales"]])
```

```python
# Dropping Duplicates 
store_types = sales.drop_duplicates(subset=["store", "type"])
store_types.head()
```

```python
store_depts = sales.drop_duplicates(subset=["store", "department"])
store_depts.head()
```

```python
# Calc total weekly sales
sales_all = sales["weekly_sales"].sum()
 
# Subset for type A stores, calc total weekly sales
sales_A = sales[sales["type"] == "A"]["weekly_sales"].sum()
 
# Subset for type B stores, calc total weekly sales
sales_B = sales[sales["type"] == "B"]["weekly_sales"].sum()

# Subset for type C stores, calc total weekly sales
sales_C = sales[sales["type"] == "C"]["weekly_sales"].sum()
 
# Get proportion for each type
sales_propn_by_type = [sales_A, sales_B, sales_C] / sales_all
print(sales_propn_by_type)
```

```python
# Import numpy with the alias np
import numpy as np

# For each store type, aggregate weekly_sales: get min, max, mean, and median
sales_stats = sales.groupby('type')['weekly_sales'].agg([min, max, np.mean, np.median])

# Print sales_stats
print(sales_stats)

# For each store type, aggregate unemployment and fuel_price_usd_per_l: get min, max, mean, and median
unemp_fuel_stats = sales.groupby('type')[['unemployment','fuel_price_usd_per_l']].agg([min, max, np.mean, np.median])

# Print unemp_fuel_stats
unemp_fuel_stats
```

```python
avocados = pd.read_pickle('avoplotto.pkl')
```

```python
# Import matplotlib.pyplot with alias plt
import matplotlib.pyplot as plt

# Look at the first few rows of data
print(avocados.head())

# Get the total number of avocados sold of each size
nb_sold_by_size = avocados.groupby('size')['nb_sold'].sum()

# Create a bar plot of the number of avocados sold by size
nb_sold_by_size.plot(kind = 'bar')

# Show the plot
plt.show()
```

```python
# Get the total number of avocados sold on each date
nb_sold_by_date = avocados.groupby('date')['nb_sold'].sum()

# Create a line plot of the number of avocados sold by date
nb_sold_by_date.plot(kind='line', rot = 45)

# Show the plot
plt.show()
```

```python
# Scatter plot of avg_price vs. nb_sold with title

avocados.plot(x='nb_sold', y='avg_price', kind = 'scatter', title = 'Number of avocados sold vs. average price')

# Show the plot
plt.show()
```

```python
# Histogram of conventional avg_price
avocados[avocados['type'] == 'conventional']['avg_price'].plot(kind = 'hist')

# Histogram of organic avg_price
avocados[avocados['type'] == 'organic']['avg_price'].plot(kind = 'hist')

# Add a legend
plt.legend(['conventional','organic'])

# Show the plot
plt.show()
```

```python
# Which countrys GDP is growing during the Year 2010 and Year 2018?
import pandas as pd

df = pd.read_csv('WorldBank_GDP.csv')
df = df.drop_duplicates()
df['Year'] = df['Year'].astype(int)
df_2010 = df[df['Year'] == 2010][['Country Name', 'GDP']].set_index('Country Name')['GDP']
df_2018 = df[df['Year'] == 2018][['Country Name', 'GDP']].set_index('Country Name')['GDP']
growing = df_2010[df_2010 < df_2018].index.to_list()
print("Growing countries:", growing)

# USA GDP Growth
gdp_df = pd.read_csv('WorldBank_GDP.csv').drop_duplicates()
gdp_df[gdp_df['Country Name'] == 'United States'].plot(x='Year', y='GDP', kind='line', marker='o', figsize=(8, 5))
plt.title('USA GDP Growth')
plt.xlabel('Year')
plt.ylabel('GDP (current US$)')
plt.grid(True)
plt.tight_layout()
plt.show()

# Germany GDP Growth
gdp_df = pd.read_csv('WorldBank_GDP.csv').drop_duplicates()
gdp_df[gdp_df['Country Name'] == 'Germany'].plot(x='Year', y='GDP', kind='line', marker='o', figsize=(8, 5))
plt.title('Germany GDP Growth')
plt.xlabel('Year')
plt.ylabel('GDP (current US$)')
plt.grid(True)
plt.tight_layout()
plt.show()

# China GDP Growth
gdp_df = pd.read_csv('WorldBank_GDP.csv').drop_duplicates()
gdp_df[gdp_df['Country Name'] == 'China'].plot(x='Year', y='GDP', kind='line', marker='o', figsize=(8, 5))
plt.title('China GDP Growth')
plt.xlabel('Year')
plt.ylabel('GDP (current US$)')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Country that has the highest average temperature
df = pd.read_csv('temperatures.csv')
country_avg = df.groupby('country')['avg_temp_c'].mean()
max_country = country_avg.idxmax()
max_temp = country_avg.max()
print(f"The country with the highest average temperature is {max_country} with an average temperature of {max_temp:.2f} C.")
```

```python
# Countries with average temperature between 20 and 30 degrees celsius
df = pd.read_csv('temperatures.csv')
country_avg = df.groupby('country')['avg_temp_c'].mean()
warm = country_avg[(country_avg >= 20) & (country_avg <= 30)]
print(f"There are {len(warm)} countries with average temperature between 20 and 30 degrees Celsius: ")
warm
```
